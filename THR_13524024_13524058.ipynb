### Implementasi Manual

Data Numerik

```python
import pandas as pd
import numpy as np
from IPython.display import display

def calculate_mean(data):
    return sum(data) / len(data)

def calculate_median(data):
    n = len(data)
    sorted_data = sorted(data)
    if n % 2 == 1:
        return sorted_data[n // 2]
    else:
        return (sorted_data[n // 2 - 1] + sorted_data[n // 2]) / 2

def calculate_mode(data):
    counts = {}
    for value in data:
        counts[value] = counts.get(value, 0) + 1
    if not counts: return None
    max_count = max(counts.values())
    modes = [key for key, value in counts.items() if value == max_count]
    return modes[0]

def calculate_variance(data, ddof=1):
    n = len(data)
    if n <= ddof: return 0
    mean = calculate_mean(data)
    return sum((x - mean) ** 2 for x in data) / (n - ddof)

def calculate_std_dev(data, ddof=1):
    return calculate_variance(data, ddof) ** 0.5

def calculate_quartiles(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    def quantile(p):
        index = (n - 1) * p
        i = int(index)
        f = index - i
        if i >= n - 1: return sorted_data[-1]
        return sorted_data[i] + f * (sorted_data[i+1] - sorted_data[i])
    Q1 = quantile(0.25)
    Q3 = quantile(0.75)
    IQR = Q3 - Q1
    return Q1, Q3, IQR

def calculate_skewness(data):
    n = len(data)
    if n < 3: return 0
    mean = calculate_mean(data)
    var_pop = sum((x - mean) ** 2 for x in data) / n
    std_pop = var_pop ** 0.5
    if std_pop == 0: return 0
    
    moment3 = sum((x - mean) ** 3 for x in data) / n
    skew_pop = moment3 / (std_pop ** 3)

    return skew_pop * (n * (n - 1)) ** 0.5 / (n - 2)

def calculate_kurtosis(data):
    n = len(data)
    if n < 4: return 0
    mean = calculate_mean(data)
    var_pop = sum((x - mean) ** 2 for x in data) / n
    std_pop = var_pop ** 0.5
    if std_pop == 0: return 0
    
    moment4 = sum((x - mean) ** 4 for x in data) / n
    kurt_pop = moment4 / (std_pop ** 4)
    
    factor1 = (n + 1) * (n - 1) / ((n - 2) * (n - 3))
    factor2 = 3 * ((n - 1) ** 2) / ((n - 2) * (n - 3))
    return factor1 * kurt_pop - factor2

df = pd.read_csv("wine.csv")
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

stats_functions = {
    'Mean': calculate_mean, 'Median': calculate_median, 'Modus': calculate_mode,
    'Standar Deviasi': calculate_std_dev, 'Variansi': calculate_variance,
    'Skewness': calculate_skewness, 'Kurtosis': calculate_kurtosis
}

results = {}
for col in numerical_cols:
    data = df[col]
    res = {name: func(data) for name, func in stats_functions.items()}
    res['Min'] = min(data)
    res['Max'] = max(data)
    res['Range'] = res['Max'] - res['Min']
    Q1, Q3, IQR = calculate_quartiles(data)
    res['Q1'], res['Q3'], res['IQR'] = Q1, Q3, IQR
    results[col] = res

df_results = pd.DataFrame(results).T
order = ['Mean', 'Median', 'Modus', 'Standar Deviasi', 'Variansi', 'Range', 
            'Min', 'Max', 'Q1', 'Q3', 'IQR', 'Skewness', 'Kurtosis']
display(df_results[order])
```

Data Kategorikal

```python
import pandas as pd
from IPython.display import display

def analyze_categorical_manual(series):
    counts = {}
    total_count = 0
    for item in series:
        counts[item] = counts.get(item, 0) + 1
        total_count += 1
        
    unique_values = list(counts.keys())
    results_list = []
    for value, freq in counts.items():
        percent = (freq / total_count) * 100
        results_list.append({
            'Kategori': value,
            'Frekuensi': freq,
            'Persentase (%)': percent
        })
    return unique_values, results_list

df = pd.read_csv("wine.csv")
categorical_cols = df.select_dtypes(include=['object', 'category']).columns.tolist()

for col in categorical_cols:
    print(f"\nKolom: '{col}'")
    vals, stats = analyze_categorical_manual(df[col])
    print(f"Nilai Unik: {vals}")
    
    df_cat = pd.DataFrame(stats).sort_values(by='Frekuensi', ascending=False)
    display(df_cat)
```

## Nomor 2
Apakah pada data tersebut terdapat outlier? Jika ada, tangani outlier tersebut dan jelaskan cara penanganannya

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

def calculate_quartiles(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    
    def quantile(p):
        index = (n - 1) * p
        i = int(index)
        f = index - i
        if i >= n - 1: return sorted_data[-1]
        return sorted_data[i] + f * (sorted_data[i+1] - sorted_data[i])
            
    Q1 = quantile(0.25)
    Q3 = quantile(0.75)
    IQR = Q3 - Q1
    return Q1, Q3, IQR

def detect_outliers_manual(data):
    Q1, Q3, IQR = calculate_quartiles(data)
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    outliers = [x for x in data if x < lower_bound or x > upper_bound]
    return outliers, lower_bound, upper_bound

df = pd.read_csv("wine.csv")
numerical_cols = df.select_dtypes(include=np.number).columns.tolist()

cols_with_outliers = []
outlier_summary = []

for col in numerical_cols:
    data = df[col]
    outliers, lower, upper = detect_outliers_manual(data)
    num_outliers = len(outliers)
    
    if num_outliers > 0:
        cols_with_outliers.append(col)
        pct = (num_outliers / len(df)) * 100
        outlier_summary.append({
            'Kolom': col,
            'Jumlah Outlier': num_outliers,
            'Persentase (%)': f"{pct:.2f}%",
            'Batas Bawah': f"{lower:.4f}",
            'Batas Atas': f"{upper:.4f}"
        })

if outlier_summary:
    df_summary = pd.DataFrame(outlier_summary)
    display(df_summary)
else:
    print("Tidak ditemukan outlier pada dataset.")

plt.figure(figsize=(15, 10))
plt.suptitle('Box Plot Outlier', fontsize=16)

n_cols_plot = 3
n_rows_plot = (len(numerical_cols) + n_cols_plot - 1) // n_cols_plot

for i, col in enumerate(numerical_cols):
    plt.subplot(n_rows_plot, n_cols_plot, i + 1)
    sns.boxplot(y=df[col], color='skyblue')
    plt.title(col, fontsize=10)
    plt.ylabel('')
    
plt.tight_layout()
plt.show()
```

Pada kode di atas, pengecekan dilakukan menggunakan metode Interquartile Range (IQR) dengan langkah sebagai berikut:

1. Hitung Kuartil: Data diurutkan (sorted) lalu dicari nilai Kuartil 1 (Q1) (posisi 25%) dan Kuartil 3 (Q3) (posisi 75%) menggunakan rumus interpolasi linear manual.
2. Hitung IQR: IQR = Q3 − Q1.
3. Tentukan Batas (Fence): Batas Bawah (Lower Bound) = Q1−(1.5×IQR) dan Batas Atas (Upper Bound) = Q3+(1.5×IQR)
4. Deteksi: Setiap data (x) diperiksa. Jika x < Batas Bawah atau x > Batas Atas, maka data tersebut dikategorikan sebagai outlier.

Di data wine.csv, algoritma mendeteksi adanya outlier pada kolom:
1. chlorides: Terdapat 16 data ekstrem.
2. pH: Terdapat 7 data ekstrem.
3. sulphates: Terdapat 2 data ekstrem.

Ini terlihat jelas pada Box Plot yang dihasilkan, di mana terdapat titik-titik hitam (diamonds) yang berada di luar "kumis" (whiskers) kotak pada kolom-kolom tersebut.

Ketika outlier terdeteksi, ada 3 tindakan umum yang biasa dilakukan tergantung pada konteks datanya:
1. Dihapus (Trimming/Dropping): Menghapus seluruh baris data yang mengandung outlier
2. Dibatasi (Capping/Winsorizing): Mengganti nilai outlier dengan nilai ambang batas maksimum atau minimum (Batas Atas/Bawah IQR)
3. Transformasi Data: Mengubah skala seluruh data menggunakan fungsi matematika, misalnya Logaritma (log(x)) atau Akar Kuadrat (x​)

### Penanganan Outlier

Metode penanganan outlier yang digunakan kali ini adalah Capping. Capping adalah teknik dalam statistika dan pembersihan data untuk menangani outlier dengan cara membatasi nilai maksimum dan minimum data tersebut, alih-alih menghapusnya. Alasan mengapa digunakan metode capping:
1. Mencegah Kehilangan Informasi
2. Menstabilkan Statistik Deskriptif
3. Menangani Natural Outlier

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display

def calculate_quartiles(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    def quantile(p):
        index = (n - 1) * p
        i = int(index)
        f = index - i
        if i >= n - 1: return sorted_data[-1]
        return sorted_data[i] + f * (sorted_data[i+1] - sorted_data[i])    
    Q1 = quantile(0.25)
    Q3 = quantile(0.75)
    IQR = Q3 - Q1
    return Q1, Q3, IQR

def cap_outliers(df, column):
    Q1, Q3, IQR = calculate_quartiles(df[column])
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    def limit_value(x):
        if x < lower_bound:
            return lower_bound  
        elif x > upper_bound:
            return upper_bound  
        else:
            return x    
    df[column] = df[column].apply(limit_value)
    return df, lower_bound, upper_bound


df = pd.read_csv("wine.csv")
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
df_capped = df.copy()

cols_with_outliers = []

for col in numerical_cols:
    Q1, Q3, IQR = calculate_quartiles(df[col])
    lb_check = Q1 - 1.5 * IQR
    ub_check = Q3 + 1.5 * IQR
    
    outliers = [x for x in df[col] if x < lb_check or x > ub_check]
    
    if len(outliers) > 0:
        cols_with_outliers.append(col)
        
        df_capped, lb, ub = cap_outliers(df_capped, col)

num_plot = len(cols_with_outliers)
plt.figure(figsize=(12, 5 * num_plot))

for i, col in enumerate(cols_with_outliers):
    plt.subplot(num_plot, 2, 2*i + 1)
    sns.boxplot(x=df[col], color='salmon')
    plt.title(f"BEFORE: {col} (Data Asli)")
    plt.xlabel("")
    
    plt.subplot(num_plot, 2, 2*i + 2)
    sns.boxplot(x=df_capped[col], color='lightgreen')
    plt.title(f"AFTER: {col} (Setelah Capping Manual)")
    plt.xlabel("")

plt.tight_layout()
plt.show()

mean_asli = df[cols_with_outliers].mean()
mean_capped = df_capped[cols_with_outliers].mean()

comparison = pd.DataFrame({
    'Mean Asli': mean_asli,
    'Mean Capped': mean_capped,
    'Selisih': mean_capped - mean_asli
})
display(comparison)
```
